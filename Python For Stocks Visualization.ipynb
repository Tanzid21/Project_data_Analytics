```python

# Define your stock symbol and date range
stock_symbol = 'SOFI'  # Replace with your stock symbol
end_date = dt.datetime.now()
start_date_36mo = end_date - pd.DateOffset(months=12)
start_date_6mo = end_date - pd.DateOffset(months=6)

# # Fetch the stock data
df = pdr.get_data_yahoo(stock_symbol, start_date_36mo, end_date)

# Ensure the DataFrame index is a DatetimeIndex and is sorted
df = df.sort_index()

# Find the minimum closing price of the last 36 months
min_value_36mo = df.loc[start_date_36mo:end_date, 'Close'].min()

# Find the maximum closing price of the last 6 months
max_value_6mo = df.loc[start_date_6mo:end_date, 'Close'].max()

print(f"Minimum Close value in the last 36 months: {min_value_36mo}")
print(f"Maximum Close value in the last 6 months: {max_value_6mo}")
```
